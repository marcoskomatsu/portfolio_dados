```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros iniciais
S0 = 100          # Preço inicial da ação
mu = 0.10         # Retorno anual esperado (10%)
sigma = 0.2       # Volatilidade anual (20%)
T = 1             # Tempo em anos
n_simulacoes = 1000
n_passos = 252    # 252 dias úteis no ano

# Gerando caminhos aleatórios
dt = T / n_passos
simulacoes = np.zeros((n_passos, n_simulacoes))
simulacoes[0] = S0

for t in range(1, n_passos):
    z = np.random.standard_normal(n_simulacoes)
    simulacoes[t] = simulacoes[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)

# Plotando algumas simulações
plt.figure(figsize=(10, 6))
for i in range(10):  # Mostra só 10 caminhos
    plt.plot(simulacoes[:, i], lw=1)

plt.title('Simulação de Preço de Ação - Monte Carlo')
plt.xlabel('Dias')
plt.ylabel('Preço da Ação')
plt.grid(True)
plt.show()
```
